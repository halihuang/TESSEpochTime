```python
from preprocess import preprocess_ztf_tess, display_all_passbands, format_title, find_max_light
import os
import warnings
import matplotlib.pyplot as plt

params = {
        "norm": True,
        "to_bin": True,
        "bin_interval": "0.5D",
        "time_scale": "trigger",
        "ztf_tess": True,
        "sub_bg_model": False,
        "remove_extinction": True
    }
ztf_data = os.listdir("TESS_data/ztf_data/")
test = "2018gku_ZTF18abwoxal_detections.csv"
# 2021yyh_ZTF18acdyyxk_detections.csv"

def plot_curve(filename, save=False, save_dir="./TESS_data/ztf_tess_plots/", save_name="test"):
    xlim = [-60,60]
    fig, axs = plt.subplots(1, 2, figsize=(12, 3))

    # plot normalized
    df, meta = preprocess_ztf_tess(filename, params)
    if df is not None:
        display_all_passbands(df, meta, flip=True, xlim=xlim, ax=axs[0], title_info="Normalized")

        # plot unnormalized
        params['norm'] = False
        df_nonorm, meta = preprocess_ztf_tess(filename, params)
        display_all_passbands(df_nonorm, meta, flip=True, xlim=xlim, ax=axs[1], title_info="Unnormalized")

        fig.suptitle(format_title(meta), y=1.35)
        if save:
            fig.savefig(save_dir + save_name, bbox_inches="tight")

    plt.close(fig)

# plot_curve(test, save=True, save_name="file")

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for file in ztf_data:
            name = file.split(".")[0]
            print(name)
            plot_curve(file, save=True, save_name=name)
```
